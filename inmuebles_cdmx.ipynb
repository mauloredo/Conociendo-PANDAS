Pandas ofrece diversas funciones para importar y exportar datos en diferentes formatos. Las principales funciones de Pandas para importar datos son las siguientes:

read_csv(): Esta función se utiliza para leer archivos CSV (Comma Separated Values), que son archivos de texto que contienen datos separados por comas. Es posible proporcionar varios parámetros para personalizar la lectura del CSV, como el delimitador, el encabezado, el tipo de codificación, entre otros.

read_excel(): Esta función se utiliza para leer archivos de Excel (.xls o .xlsx) y crear un DataFrame a partir de los datos.

read_json(): Esta función se utiliza para leer archivos JSON (JavaScript Object Notation), que son archivos de texto que contienen datos en formato de objeto JavaScript.

read_html(): Esta función se utiliza para leer tablas HTML, que son estructuras de datos organizadas en formato de tabla en una página web.

read_sql(): Esta función se utiliza para leer datos de una base de datos relacional, como MySQL, PostgreSQL y SQL Server. Pandas es capaz de importar datos de diferentes maneras, lo que permite ajustar parámetros como la consulta, el nombre de la tabla y el tipo de datos de las columnas.

# **Demandas de ML**

**Valores promedio de alquiler por tipo de inmueble**
Para hacer este sencillo cálculo, es necesario hacer grupos en en DF para despues calcular el promedio.

```python
datos.groupby("Tipo")["Valor"].mean()
```

Para modificar el resultado, con características de DF y en valor ascendente de acorde al valor, se ejecuta lo siguiente:

```python
datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

IMPORTANTE: Para almacenar consultas al DF como lo acabamos de hacer, se utiliza el siguiente código:

```python
df_Tipo_Valor = datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

Para graficar esta consulta:

```python
df_Tipo_Valor.plot(kind="barh", figsize=(12,8), color='purple')
```

**Porcentaje de cada tipo de inmueble presente en la base de datos**

1. Identificar los tipos de propiedad

```python
datos["Tipo"].unique()
```

2. Clasificar las propiedades

```python
comerciales = [
    "Conjunto Comercial/Sala", "Edificio Completo",
    "Tienda/Salón", "Casa Comercial", "Galpón/Depósito/Almacén",
    "Tienda en Centro Comercial", "Hotel", "Industria"
]

no_comerciales = [
    "Cocineta", "Casa", "Departamento", "Casa de Condominio",
    "Casa de Vecindad", "Terreno Estándar", "Cochera/Estacionamiento",
    "Loft", "Rancho", "Posada/Chalé", "Estudio", "Loteo/Condominio", "Departamento en Hotel"
]

```

3. Agrupamient de propiedade en PANDAS

```python
# Filtrar propiedades comerciales
datos_comerciales = datos[datos["Tipo"].isin(comerciales)]

# Filtrar propiedades no comerciales
datos_no_comerciales = datos[datos["Tipo"].isin(no_comerciales)]
```

4. Conta las propiedades por tipo

```python
conteo_comercial = datos_comerciales["Tipo"].value_counts()
conteo_no_comercial = datos_no_comerciales["Tipo"].value_counts()
```

```python
conteo_comercial
```

```python
conteo_no_comercial
```

5. Calcular el porcentaje dentro de cada grupo

```python
porcentaje_comercial = (conteo_comercial / conteo_comercial.sum()) * 100
porcentaje_no_comercial = (conteo_no_comercial / conteo_no_comercial.sum()) * 100
```

```python
porcentaje_comercial
```

```python
porcentaje_no_comercial
```

6. Graficar resultados

```python
import matplotlib.pyplot as plt

# Configuración de gráficos
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Propiedades comerciales
porcentaje_comercial.plot(kind="barh", ax=axes[0], color="orange")
axes[0].set_title("Distribución de Propiedades Comerciales (%)")
axes[0].set_xlabel("Porcentaje")

# Propiedades no comerciales
porcentaje_no_comercial.plot(kind="barh", ax=axes[1], color="blue")
axes[1].set_title("Distribución de Propiedades No Comerciales (%)")
axes[1].set_xlabel("Porcentaje")

plt.tight_layout()
plt.show()
```

# **Group By**

by: Este es el parámetro más común y se utiliza para especificar la columna o columnas por las cuales queremos agrupar. Como argumento, podemos pasar el nombre de una columna o una lista de nombres de columnas;

axis: Utilizamos este parámetro para especificar el eje a lo largo del cual queremos agrupar. El valor predeterminado es 0, lo que significa que se agruparán las filas. Si deseamos agrupar las columnas, debemos establecer este parámetro en 1;

sort: Este parámetro es un booleano (True o False) que indica si los grupos deben ordenarse por el valor de la columna de agrupación. El valor predeterminado es True;

dropna: Utilizamos este parámetro para controlar si los valores faltantes (NaN) se eliminarán o no durante el proceso de agrupación. El valor predeterminado es True.
