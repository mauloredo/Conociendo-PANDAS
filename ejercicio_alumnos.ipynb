# ELABORACIÓN

Los estudiantes participaron en una actividad extracurricular y ganaron puntos extras. Estos puntos extras corresponden al 40% de su nota actual. Por lo tanto, crea una columna llamada "Puntos_extras" que contenga los puntos extras de cada estudiante, es decir, el 40% de su nota actual.

```python

# Identificar valores NaN en el DataFrame
nan_counts = df.isnull().sum()

# Mostrar el conteo de valores NaN por columna
print(nan_counts)
```

```python
#reemplazar nans por 0
df.fillna(0, inplace=True)

# Verificar que los cambios se hayan aplicado
print(df.head())
```

```python
df
```

```python
# Crear nueva columna con el nombre Extras. el valor de cada fila para esta columna es la multiplicación de la calificación por .4

df["Extras"] = df["Nota"] * 0.4


df
```

```python
# Crear la columna "Final" sumando "Nota" y "Extras"
df["Final"] = df["Nota"] + df["Extras"]

# Limitar los valores de la columna "Final" a un máximo de 10
df["Final"] = df["Final"].clip(upper=10)


df
```

```python
# Cambiar el nombre de la columna "Aprobado" por "Preliminar"
df.rename(columns={"Aprobado": "Preliminar"}, inplace=True)

df
```

```python
# Crear la columna "Aprobado" con valores booleanos
df["Aprobado"] = df["Final"] >= 7

df
```
